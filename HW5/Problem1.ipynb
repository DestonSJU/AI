## **Function Setup**
Next I set up the function that will be used for the genetic algorithm. To do this I use α and β to weight the f1(x) and f2(x) functions respectively and then add them together in order to get 1 f(x). I initially give both functions an equal weight of 0.5. I then set up the x1, x2, x3, x4 parameters.

```r
#Create α and β variables
alpha <- 0.5
beta <- 0.5

#Create the f(x) = α*f1(x) + β*f2(x) function
#f1(x) = 2 * x2 * x4 + x3 * (x1 - 2 * x4)
#f2(x) = 60000 / (x3 * ((x1 - 2 * x4) ** 3) + 2 * x2 * x4 * (4 * (x4 ** 2) + 3 * x1 * (x1 - 2 * x4)))

func <- function(x1, x2, x3, x4) {
    alpha * (2 * x2 * x4 + x3 * (x1 - 2 * x4)) + beta *
     (60000 / (x3 * ((x1 - 2 * x4) ** 3) + 2 * x2 * x4 * (4 * (x4 ** 2) + 3 * x1 * (x1 - 2 * x4))))
}
```

```r
#Set up parameters for the function
x1 <- seq(10, 80, by = 0.1)
x2 <- seq(10, 50, by = 0.1)
x3 <- seq(0.9, 5.0, by = 0.1)
x4 <- seq(0.9, 5.0, by = 0.1)
```

## **GA with Equal Weight**
Next I run the genetic algorithm, displaying the graph of the best and average fitness value in each generation, and the best solution for each generation. After looking at the data, I noticed that the best and average fitness values converged at around the 40th generation. I found that the overall best solution (13.01451,	11.22062,	0.983265,	1.401321) trended towards minimizing f1 over f2 since the x values in the solution are closer to the best values for f1 than f2.

```r
#Call the ga function to create the genetic algorithm
GA <- ga(type = "real-valued",
         fitness =  function(x) -func(x[1], x[2], x[3], x[4]),
         lower = c(10, 10, 0.9, 0.9), upper = c(80, 50, 5.0, 5.0),
         popSize = 50, pcrossover = 0.75, pmutation = 0.001, maxiter = 100, run = 50, keepBest = TRUE, monitor = TRUE, seed = 100)

#Displays the best solution for each generation
GA@bestSol

#Displays information about the genetic algorithm
summary(GA)

#Displays the graph for the best and average fitness for each generation
plot(GA)
```
